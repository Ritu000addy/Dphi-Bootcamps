```python
df1["Regional indicator"].unique()
```

```python
Generosity_East_Asian_countries = df1[df1["Regional indicator"] == "East Asia" ]["Generosity"]
print(Generosity_East_Asian_countries)
```

```python
Ladder_score = df1[df1["Regional indicator"] == "East Asia"][ "Ladder score"]
print(Ladder_score)
```

```python
plt.plot(Generosity_East_Asian_countries, Ladder_score)
plt.xlabel("Generosity")
plt.ylabel("Ladder Score")
plt.title("Line Plot")
plt.show()
```

```python
Ladder_score_SEA = df1[df1["Regional indicator"] == "Southeast Asia"]["Ladder score"].head(5)
print(Ladder_score_SEA)
```

```python
Countries = df1[df1["Regional indicator"] == "Southeast Asia"]["Country name"].head(5)
print(Countries)
```

```python
plt.bar( Countries, Ladder_score_SEA)
plt.xlabel("Country name")
plt.ylabel("Ladder Score")
plt.title("Variation in South East Asia")
plt.show()
```

```python
df_2021 = df1[df1["Country name"].isin(['China', 'Nepal','Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'])]
df_2021.head()
```

```python
plt.scatter(df_2021["Country name"], df_2021["Logged GDP per capita"])
plt.xlabel("Country name")
plt.ylabel("Logged GDP per capita")
plt.title("Scatter plot")
plt.show()
```
